  # Question Three: Missing Number in Range
 
  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [0, 2]`

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

  Output: [2, 3, 4]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 0, 1, 10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
from typing import List

def missing_num(nums: List[int]):
    """Return list of numbers missing from range [0, n] where n = max(nums). Return -1 if none missing."""
    if not nums:
        return -1
    n = max(nums)
    seen = set(nums)
    missing = [i for i in range(n + 1) if i not in seen]
    return -1 if not missing else missing

# Test with provided examples:
print(missing_num([0, 2]))                           # [1]
print(missing_num([5, 0, 1]))                        # [2, 3, 4]
print(missing_num([6, 8, 2, 3, 5, 7, 0, 1, 10]))    # [4, 9]
```

```python
# Paraphrase: Valid Bracket Sequence (Partner's Problem - Question 2)
# Given a string of only bracket characters (), [], {}, determine if every opening bracket
# is properly closed by the correct type of bracket in the correct order. Valid means
# pairs must match (e.g., '(' with ')') and nesting must be correct (e.g., "([)]" is invalid).
```


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
# New Example ( demonstrates understanding ):
# Input:  s = "()"
# Output: True  (single pair, correctly closed)

# Trace of partner's Example 1: Input: s = "([]{})"
# Step 1: '(' â†’ opening, push to stack: ['(']
# Step 2: '[' â†’ opening, push: ['(', '[']
# Step 3: ']' â†’ closing, pop '[' matches ']' âœ“, stack: ['(']
# Step 4: '{' â†’ opening, push: ['(', '{']
# Step 5: '}' â†’ closing, pop '{' matches '}' âœ“, stack: ['(']
# Step 6: ')' â†’ closing, pop '(' matches ')' âœ“, stack: []
# Final: stack empty â†’ return True. All brackets properly paired in correct order.
```


-   Copy the solution your partner wrote. 


```python
def is_valid_brackets(s: str) -> bool:
    # define the bracket map
    bracket_map = {')': '(', ']': '[', '}': '{'} 
    
    # use stack as described in the lesson
    stack = []

    for char in s:
        # check if a closer bracket
        if char in bracket_map:
            # Pop top pringles if stack isn't empty, else use a dummy value '#'
            top_element = stack.pop() if stack else '#'
            
            # If the opener does not match the closer
            if bracket_map[char] != top_element:
                return False
        else:
            # An opening bracket, wait for its partner
            stack.append(char)

    # Done? Stack empty? 
    # not stack true, partner found
    return not stack
```


-   Explain why their solution works in your own words.


```python
# Why the solution works:
# A stack is ideal because brackets must be closed in reverse order of opening (LIFO).
# For each character: if it's a closing bracket, the most recently opened bracket (stack top)
# must be its matching pairâ€”otherwise the sequence is invalid. If it's an opening bracket,
# we push it and wait for its closer. At the end, the stack must be empty (all pairs closed).
```


-   Explain the problemâ€™s time and space complexity in your own words.


```python
# Time complexity: O(n) â€” single pass over the string; each push/pop is O(1).
# Space complexity: O(n) â€” worst case all opening brackets (e.g., "((("), stack holds all.
```
